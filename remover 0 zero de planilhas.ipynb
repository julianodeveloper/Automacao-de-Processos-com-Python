```python
import pandas as pd

# Função para remover linhas com valor 0
def remover_linhas_com_zero(caminho_arquivo, caminho_saida):
    """
    Remove linhas que contêm o valor 0 em qualquer coluna de um arquivo Excel.

    Parâmetros:
        caminho_arquivo (str): Caminho do arquivo Excel de entrada.
        caminho_saida (str): Caminho do arquivo Excel de saída.
    """
    # Ler o arquivo Excel
    if caminho_arquivo.endswith('.xls'):
        df = pd.read_excel(caminho_arquivo, engine='xlrd')  # Para arquivos .xls
    else:
        df = pd.read_excel(caminho_arquivo)  # Para arquivos .xlsx

    # Remover linhas que contêm 0 em qualquer coluna
    df_sem_zero = df[~(df == 0).any(axis=1)]

    # Salvar o resultado em um novo arquivo Excel
    df_sem_zero.to_excel(caminho_saida, index=False)
    print(f"Arquivo salvo em: {caminho_saida}")

# Caminho do arquivo de entrada e saída
caminho_arquivo = 'seu_arquivo.xls'  # Substitua pelo caminho do seu arquivo
caminho_saida = 'arquivo_sem_zero.xlsx'  # Nome do arquivo de saída

# Executar a função
remover_linhas_com_zero(caminho_arquivo, caminho_saida)
```

Observações:
O código funciona para arquivos .xls e .xlsx.

Se o arquivo de saída for .xlsx, ele será criado usando o engine openpyxl por padrão.
