```python
import pandas as pd
import matplotlib.pyplot as plt
from fpdf import FPDF

# Simulando dados econômicos
data = {
    "Indicador": ["PIB", "Inflação", "Juros", "Câmbio"],
    "Valor": [2.5, 4.3, 11.75, 5.10]
}

df = pd.DataFrame(data)

# Criando um gráfico
plt.figure(figsize=(6,4))
plt.bar(df["Indicador"], df["Valor"], color=['blue', 'red', 'green', 'purple'])
plt.ylabel("Valor (%)")
plt.title("Indicadores Econômicos")
plt.savefig("grafico.png")
plt.close()

# Criando um PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", style="B", size=16)
pdf.cell(200, 10, "Relatório Econômico", ln=True, align="C")

# Adicionando tabela
pdf.set_font("Arial", size=12)
for i in range(len(df)):
    pdf.cell(50, 10, df.iloc[i, 0], border=1)
    pdf.cell(50, 10, str(df.iloc[i, 1]), border=1)
    pdf.ln()

# Adicionando gráfico
pdf.image("grafico.png", x=50, y=None, w=100)




# Salvando relatório
pdf.output("relatorio.pdf")

print("Relatório gerado com sucesso!")
```
